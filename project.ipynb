#  STEP-2 (Data Preprocessing and Analysis):

2.1 Preprocessing (Drop "Name", Scale features)

```python
data = data.dropna() 
print(data)
```

```python
data = data.fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler

# Drop the "Name" column before scaling
data_numeric = data.drop("Name", axis=1)

# Apply scaling only to numeric features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_numeric)

# Convert scaled array back into a DataFrame
data_scaled = pd.DataFrame(scaled_features, columns=data_numeric.columns)

print("\n Scaled Data:")
print(data_scaled)
```

2.2 : Univariate Analysis

```python
for col in data_scaled.columns:
    sns.histplot(data_scaled[col], kde=True)      # Creates a histogram  using Seaborn (sns) and kde=True adds a smooth curve
    plt.title(f'Distribution of {col}')           # Adds a title to each plot, showing which column you're visualizing.
    plt.show()                                    # 
```

```python
# if we want to specify any col then code will be:

column_name = 'Murder'

sns.histplot(data_scaled[column_name], kde=True)
plt.title(f'Distribution of {column_name}')
plt.show()
```

2.3 : Bivariate Analysis

```python

sns.heatmap(data_scaled.corr(), annot=True, cmap='coolwarm')    
plt.title('Correlation Matrix')
plt.show()
```

# STEP-3 (Data splitting) :

Create Target Variable (Binary Classification):

```python
data_scaled['Crime_level'] = ['High' if x > data_scaled['Murder'].median()else 'Low' for x in data_scaled['Murder']]
x = data_scaled.drop('Crime_level',axis=1)
y = data_scaled['Crime_level']
```

Train-Test Split:

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train distribution:\n", y_train.value_counts())
print("y_test distribution:\n", y_test.value_counts())

```
